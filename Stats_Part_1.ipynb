## Measure Of Dispersion

1. Variance
2. Standard Deviation

```python
statistics.variance(ages)
```

```python
np.var(ages,axis=0, ddof=1) # ddof=0 show sample variance and dd0f=1 show population variance
```

```python
def variance(data):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-1)
  return variance

```

```python
variance(ages)
```

```python
def variance(data,dof=0):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

```python
variance(ages,dof=1)
```

```python
statistics.pvariance(ages)
```

```python
#standard deviation

import statistics
import numpy as np

print(statistics.stdev(ages)) # statistic tool only show population std
print("sample standard deviation is: ", np.std(ages,axis=0, ddof=0)) # ddof=0 show sample standard deviation
print("population standard deviation is: ", np.std(ages,axis=0, ddof=1)) # dd0f=1 show population standard deviation
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

## Check whether Disstribution is Normal Distribution

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
mu, sigma = 3., 1. # mean and standard deviation
sample = np.random.lognormal(mu, sigma, 100)
```

```python
def plot_data_log(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(np.log(sample),dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data_log(sample)
```
